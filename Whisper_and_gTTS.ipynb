# Load the Library Model

```python
import whisper

model = whisper.load_model("medium")
```

#Getting Audio File

```python
!wget -O audio.mp3 http://www.moviesoundclips.net/movies1/batmanbegins/bats.mp3
```

```python
from IPython.display import Audio,display

display(Audio('audio.mp3',autoplay=True))
```

#Speech to text transcription

```python
result = model.transcribe('audio.mp3')
print(result['text'])
```

#Alternate way
## It automatically creates srt and txt file for generated text

```python
!whisper 'audio.mp3' --model medium.en
```

#Using Tiny model

```python

import whisper

tiny = whisper.load_model("tiny")
```

```python
result = tiny.transcribe("audio.mp3")
print(result["text"])
```

#Google Text to speech(GTTS)

```python
!pip install gTTS
```

```python
from gtts import gTTS
from IPython.display import Audio
tts=gTTS(result['text'])
```

```python
tts.save('batman-begins.wav')
sound_file = 'batman-begins.wav'
Audio(sound_file, autoplay = True)
```
